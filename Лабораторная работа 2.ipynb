```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Данные по варианту

```python
y = [15,29,36,43,57,64,76,86,92,98,103,106]
x = [2,4,5,6,8,9,11,12,13,15,16,18]
```

График по точкам

```python
plt.plot(x,y,label = 'функция 1')
plt.xlabel("x")
plt.ylabel('y')
plt.title("Первый график")
plt.grid()
plt.legend()
```

необходимо произвести квадратичную аппроксимацию

```python
x1 = 5
x2 = 11
x3 = 15
my_dict = dict(zip(x,y))
my_dict
```

```python
a0 = my_dict[x1]
a0
```

```python
a1 = (my_dict[x2] - my_dict[x1])/(x2 - x1)
a1
```

```python
a2 = 1/(x3-x2)*((my_dict[x3]-my_dict[x1])/(x3-x1)-a1)
a2
```

Воспользуемся lambda function


```python
f = lambda x: a0 + a1*(x-x1)+a2*(x-x2)*(x-x1)

f(5)
```

```python
plt.plot(x,y,label = 'функция 1')
plt.plot(x,f(np.array(x)),label = 'функция 2')
plt.xlabel("x")
plt.ylabel('y')
plt.title("Второй  график")
plt.grid()
plt.legend()
```

```python

plt.plot(np.linspace(-1000,1000,1000),f(np.linspace(-1000,1000,1000)),label = 'функция 2')
```

Интерполирование функции


```python
a = (len(x) * sum([i*j for i,j in my_dict.items()])- sum(x)*sum(y)) /(len(x)* sum([i**2 for i in x]) - sum(x)**2)
a
```

```python
b = (sum(y) - a*sum(x))/ len(x)
b
```

```python
f2 = lambda x: a*x+b
```

```python
plt.plot(x,y,label = 'функция 1')
plt.plot(x,f(np.array(x)),label = 'функция 2')
plt.plot(x,f2(np.array(x)),label = 'функция 3')
plt.xlabel("x")
plt.ylabel('y')
plt.title("Третий  график")
plt.grid()
plt.legend()
```
